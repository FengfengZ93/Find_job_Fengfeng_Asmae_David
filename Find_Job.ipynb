```python
import pandas as pd
import json
import datetime as dt
```

```python
# Opening JSON file


df = pd.read_json('/content/drive/MyDrive/Formation_IA/data.json')

```

```python
#Remove the '/n'
df = df.replace('\n','', regex=True)
df['Intitulé du poste'] = [item[0].replace('\n', '') for item in df['Intitulé du poste']]
df['lieu'] = [', '.join(item).replace('\n', '') for item in df['lieu']]
df['competences'] = [' '.join(item).replace('\n', '').replace(' ', ', ') for item in df['competences']]
df['Type de poste'] = [''.join(item).replace('\n', ', ').replace(', , ', '').strip() for item in df['Type de poste']]

df
```

```python
# Split single column into several columns use ',' delimiter
df[['lieu', 'salary']] = df["lieu"].str.split(",", expand = True, n=1)

df[['min_salary', 'max_salary']] = df["salary"].str.split("-", expand = True, n=1)

df[['comp0','comp1', 'comp2', 'comp3', 'comp4', 'comp5']] = df["competences"].str.split(",", expand = True, n=5)

```
